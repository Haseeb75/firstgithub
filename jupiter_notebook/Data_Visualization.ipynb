```python
# step-1 import libraries
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as matplotlib
```

```python
# step-2 setting a theme
```

```python
sns.set_theme(style="ticks", color_codes= True)
```

```python
# step-4 import dataset, you can also import your own dataset
```

```python
kashti=sns.load_dataset("titanic")
```

```python
# this will give us only data not a graph
# print (kashti)
```

```python
# step-5 plot basic graph with 1 variable count
```

```python
p=sns.countplot(x="sex", data=kashti)
```

```python
# step-6 plot basic graph with 2 variables (countplot) with titles
```

```python
p=sns.countplot(x="sex", data=kashti, hue="class")
p.set_title("Titanic Count Plot")
```
